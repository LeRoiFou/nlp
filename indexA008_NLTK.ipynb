Dans ce programme, on a recours à la lemmatisation qui permet d'obtenir de meilleurs résultats de recherche, car elle réduit les différentes formes d'un mot à une forme de base commune, ce qui facilite la correspondance des différentes formes d'un mot dans le texte.

À cela :
- On assigne la fonction nltk.WordNetLemmatizer() dans un objet
- On a recours à la fonction lemmatize()

Date : 22-07-23

```python
import nltk
```

```python
# Récupération de la déclaration des droits de l'homme en anglais uniquement
# assignée dans une liste
udhr = nltk.corpus.udhr.words('English-Latin1')
```

```python
# Affichage des 30 premiers composants
udhr[:30]
```

```python
# Assignation de la fonction nltk.PorterStemmer() en objet
porter = nltk.PorterStemmer()
```

```python
# Comprehension de liste : racine des 30 premiers composants 
# de la liste ci-avant
[porter.stem(w) for w in udhr]
```

```python
#
wnl = nltk.WordNetLemmatizer()
```

```python
[wnl.lemmatize(w) for w in udhr][:30]
```
